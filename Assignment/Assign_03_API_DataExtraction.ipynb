```python
# Importing library
import requests
import pandas as pd
```

```python
# Getting the list of movie genres from API

genres_url = 'https://api.themoviedb.org/3/genre/movie/list?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US'
genres_response = requests.get(genres_url)

if genres_response.status_code == 200:     # successful request status_code=200
    genres_data = genres_response.json()
    genres = genres_data['genres']
else:
    print(f"Error {genres_response.status_code}: {genres_response.text}")
    genres = []
genres
```

```python
for i in genres:
    print(i['id'], i['name'])
```

```python
# Getting the top-rated movies data

all_movies = []
for i in range(1,501):
    url = f'https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={i}'
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        movies = data['results']
        all_movies.extend(movies)
    else:
        print(f"Error {response.status_code}: {response.text}")
        movies = []
```

```python
# Storing data in df

# df1=pd.DataFrame(all_movies)     # 200
# df1.head()
# df2=pd.DataFrame(all_movies)     # 400
# df2.head()
df3=pd.DataFrame(all_movies)     # 500
df3.head()
```

```python
# Our genre data
genre = [{'id': 28, 'name': 'Action'},
         {'id': 12, 'name': 'Adventure'},
         {'id': 16, 'name': 'Animation'},
         {'id': 35, 'name': 'Comedy'},
         {'id': 80, 'name': 'Crime'},
         {'id': 99, 'name': 'Documentary'},
         {'id': 18, 'name': 'Drama'},
         {'id': 10751, 'name': 'Family'},
         {'id': 14, 'name': 'Fantasy'},
         {'id': 36, 'name': 'History'},
         {'id': 27, 'name': 'Horror'},
         {'id': 10402, 'name': 'Music'},
         {'id': 9648, 'name': 'Mystery'},
         {'id': 10749, 'name': 'Romance'},
         {'id': 878, 'name': 'Science Fiction'},
         {'id': 10770, 'name': 'TV Movie'},
         {'id': 53, 'name': 'Thriller'},
         {'id': 10752, 'name': 'War'},
         {'id': 37, 'name': 'Western'}]

# Convert genre data to a dictionary for easy lookup
genre_dict = {g['id']: g['name'] for g in genre}

# Creating function to get genre names based on genre ids
def get_genre_names(genre_ids):
    return [genre_dict[genre_id] for genre_id in genre_ids]

# Create a new 'genre' column in the data frame
df['genre'] = df['genre_ids'].apply(get_genre_names)

df.head()
```

```python
# NOTE:
# We can not check duplicate records if data value in any of the columns is in list form

df['genre_ids'] = df['genre_ids'].apply(tuple)
df['genre'] = df['genre'].apply(tuple)

df.duplicated().sum()
```

```python
# Dropping duplicates

df=df.drop_duplicates(ignore_index=True)
df.shape
```
